```python
df = pd.read_csv("C:/동국대학교/공모전/전주공모전/구급출동_전주_2020_2022.csv", encoding='cp949')
df
```

```python
df['환자연령']
```

```python
bins = list(range(0, 101, 5)) + [float('inf')]
labels = [f'{i}~{i+4}세' for i in range(0, 100, 5)] + ['100세 이상']

# '연령상세구분' 열 생성
df['연령상세구분'] = pd.cut(df['환자연령'], bins=bins, labels=labels, right=False)
```

```python
df_year = pd.concat([pd.DataFrame(df.groupby('연령상세구분').size()).reset_index(),ratio.iloc[2:23,[2,10,18]].astype('int64').reset_index()], axis=1)
```

```python
df_year
```

```python
df_year['ratio2020'] = df_year.iloc[:,1]/df_year['2020']*100
df_year['ratio2021'] = df_year.iloc[:,1]/df_year['2021']*100
df_year['ratio2022'] = df_year.iloc[:,1]/df_year['2022']*100
```

```python
df_year
```

```python
(
    so.Plot(df_year,x='연령상세구분', y='ratio2020')
    .add(so.Bars())
)
```
